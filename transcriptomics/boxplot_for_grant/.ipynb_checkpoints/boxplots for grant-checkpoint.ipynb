# 0. user defined variables

```python
expression_file = '/home/adrian/projects/vigur/results/sequencing/tpm/DESeq2_TPM_values.tsv'
metadata_file = '/home/adrian/projects/vigur/data/sequencing/metadata/metadata.tsv'
```

# 1. read data

# 1.1. read expression

```python
expression = pandas.read_csv(expression_file, index_col=0, sep='\t')
expression.head()
```

## 1.2. read metadata

```python
metadata = pandas.read_csv(metadata_file, index_col=0, sep='\t')
metadata.head()
```

# 2. plot data

```python
samples_control_0h = metadata[(metadata['treatment'] == 'zero') & (metadata['time'] == 'zero')]['sampleID'].to_list()

samples_control_4h = metadata[(metadata['treatment'] == 'zero') & (metadata['time'] == 'four')]['sampleID'].to_list()

samples_control_24h = metadata[(metadata['treatment'] == 'zero') & (metadata['time'] == 'twentyfour')]['sampleID'].to_list()
```

```python
samples_mix_4h = metadata[(metadata['treatment'] == 'mix') & (metadata['time'] == 'four')]['sampleID'].to_list()

samples_mix_24h = metadata[(metadata['treatment'] == 'mix') & (metadata['time'] == 'twentyfour')]['sampleID'].to_list()
```

```python
samples_adr_4h = metadata[(metadata['treatment'] == 'five_epi') & (metadata['time'] == 'four')]['sampleID'].to_list()

samples_adr_24h = metadata[(metadata['treatment'] == 'five_epi') & (metadata['time'] == 'twentyfour')]['sampleID'].to_list()
```

```python
samples_nor_4h = metadata[(metadata['treatment'] == 'five_nor') & (metadata['time'] == 'four')]['sampleID'].to_list()

samples_nor_24h = metadata[(metadata['treatment'] == 'five_nor') & (metadata['time'] == 'twentyfour')]['sampleID'].to_list()
```

## 2.1. GRAMD1B

GRAMD1B is ENSG00000023171

```python
gramd1b = 'ENSG00000023171'
```

```python
c0 = expression.loc[gramd1b, samples_control_0h]
c4 = expression.loc[gramd1b, samples_control_4h]
c24 = expression.loc[gramd1b, samples_control_24h]

m4 = expression.loc[gramd1b, samples_mix_4h]
m24 = expression.loc[gramd1b, samples_mix_24h]

adr4 = expression.loc[gramd1b, samples_adr_4h]
adr24 = expression.loc[gramd1b, samples_adr_24h]

nor4 = expression.loc[gramd1b, samples_nor_4h]
nor24 = expression.loc[gramd1b, samples_nor_24h]
```

```python
my_pal = {
    0:"w", 1:"w", 2:"w",
    3:"tab:red", 4:"orange", 5:"violet",
    6:"tab:red", 7:"orange", 8:"violet"
         }

seaborn.boxplot(data=[c0, c4, c24, m4, adr4, nor4, m24, adr24, nor24], palette=my_pal)

legend_elements = [
    matplotlib.patches.Patch(facecolor='white', edgecolor='gray'),
    matplotlib.patches.Patch(facecolor='tab:red', edgecolor='gray'),
    matplotlib.patches.Patch(facecolor='orange', edgecolor='gray'),
    matplotlib.patches.Patch(facecolor='violet', edgecolor='gray')
]
matplotlib.pyplot.legend(legend_elements, ['control', 'mix', 'adr', 'nor'], ncol=2)

matplotlib.pyplot.grid(alpha=0.5, ls=':')
matplotlib.pyplot.xticks([0, 1, 2, 3, 4, 5, 6, 7, 8], ['T = 0 h', 'T = 4 h', 'T = 24 h', 'T = 4 h', 'T = 4 h', 'T = 4 h', 'T = 24 h', 'T = 24 h', 'T = 24 h'])
matplotlib.pyplot.xlabel('Conditions')
matplotlib.pyplot.ylabel('Expression (TPM)')
matplotlib.pyplot.title('GRAMD1B', style='italic')
matplotlib.pyplot.tight_layout()
```

## 2.3. heatmap of requested genes related to thyroid hormone

Requested genes: MCT8, MCT10 TRbeta1, TRalph1

```python
mct8 = 'ENSG00000147100'
mct10 = 'ENSG00000112394'
trbeta1 = 'ENSG00000126351'
tralpha1 = 'ENSG00000151090'

requested_genes = [mct8, mct10, trbeta1, tralpha1]
```

```python
ordered_samples = [
    samples_control_0h, samples_control_4h, samples_control_24h,
    samples_mix_4h, samples_adr_4h, samples_nor_4h, 
    samples_mix_24h, samples_adr_24h, samples_nor_24h
                  ]
```

```python
# build a dataframe for plotting
expression_sandbox = {}
for gene in requested_genes:
    row = []
    for samples in ordered_samples:
        sub = expression.loc[gene, samples]
        value = numpy.mean(sub)
        row.append(value)
    expression_sandbox[gene] = row
    print(gene, row[:5])
df = pandas.DataFrame.from_dict(expression_sandbox)
tdf = df.transpose()
tdf.columns = ['Control', 'Control T4', 'Control T24', 'Mix T4', 'Adr T4', 'Nor T4', 'Mix T24', 'Adr T24', 'Nor T24']
tdf.rename(index={'ENSG00000147100':'MCT8', 'ENSG00000112394':'MCT10', 'ENSG00000126351':'TRb1', 'ENSG00000151090':'TRa1'}, inplace=True)

tdf
```

```python
seaborn.set(font_scale=1.5)

seaborn.heatmap(tdf, cmap='cividis', vmin=0, vmax=25, cbar_kws={'label': 'Expression (TPM)'})
```
