```python
import pandas, matplotlib_venn

import matplotlib, matplotlib.pyplot
```

```python
input_file = 'supplemetary_information_table_SIT2.DETs.xlsx'
```

```python
half_up = pandas.read_excel(input_file, sheet_name='0.5 uM | 4 h | up', index_col=0)
half_down = pandas.read_excel(input_file, sheet_name='0.5 uM | 4 h | down', index_col=0)

five_up = pandas.read_excel(input_file, sheet_name='5 uM | 4 h | up', index_col=0)
five_down = pandas.read_excel(input_file, sheet_name='5 uM | 4 h | down', index_col=0)

fifty_up = pandas.read_excel(input_file, sheet_name='50 uM | 4 h | up', index_col=0)
fifty_down = pandas.read_excel(input_file, sheet_name='50 uM | 4 h | down', index_col=0)
```

```python
print(half_up.shape)
print(half_down.shape)
print(five_up.shape)
print(five_down.shape)
print(fifty_up.shape)
print(fifty_down.shape)

a = set(half_up.index).union(set(half_down.index))
b = set(five_up.index).union(set(five_down.index))
c = set(fifty_up.index).union(set(fifty_down.index))

print(len(a), len(b), len(c))
```

```python
matplotlib_venn.venn3(subsets=(200, 100, 50, 25, 10, 5, 1))
```

```python
font2 = {'size': 24} # use for labels
matplotlib.pyplot.rc('font', **font2) # sets the default font 

matplotlib_venn.venn3(subsets=[a, b, c],  
      set_labels=('Group A', 'Group B', 'Group C'),  
      set_colors=("tab:orange", "tab:red", "tab:purple"), alpha=0.7) 
matplotlib_venn.venn3_circles(subsets=[a, b, c],   linestyle="solid", linewidth=1) 
  
#matplotlib.pyplot.show() 
matplotlib.pyplot.savefig('venn.svg')
```
