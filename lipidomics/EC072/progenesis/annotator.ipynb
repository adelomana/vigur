```python
import pandas, os, numpy, sys
```

# user-defined variables

```python
mzdifference_threshold = 1
rtdifference_threshold = 0.5

annotations_folder = '/Users/adrian/projects/vigur/data/lipidomics/072/progenesis/annotations/'
```

```python
all_annotations = [anno0, anno1, anno2, anno3]
```

# annotate

```python
def annotation_finder(mz, rt):
    
    annotation = {}

    for annotation_df in all_annotations:
        
        lost_entries = 0
        #print('working on an annotation with {} entries.'.format(len(annotation_df)))
        for index, row in annotation_df.iterrows():
            if 'm/z' in row['Compound']:
                amz = float(row['Compound'].split('_')[1].split('m/z')[0])
                art = float(row['Compound'].split('_')[0])
                identification = row['Compound ID']
                isotope_similarity = float(row['Isotope Similarity'])
            
                # check mz
                mz_diff = numpy.abs(amz - mz)
                if mz_diff < mzdifference_threshold:
                    
                    # check RT
                    rt_diff = numpy.abs(art - rt)
                    if rt_diff < rtdifference_threshold:
                        
                        if identification not in annotation:
                            print('\t', mz, rt, mz_diff, rt_diff, identification)
                            annotation[identification] = [isotope_similarity, mz_diff]
            else:
                lost_entries = lost_entries + 1

        #print('\t lost entries (neutral mass): {}'.format(lost_entries))
        
    # sort annotation
    sorted_names = [element[0] for element in sorted(annotation.items(), key=lambda x:x[1], reverse=True)]
    string_sorted_names = ''
    for element in sorted_names:
        string_sorted_names = string_sorted_names + element + ' [{:.2f} | {:.2e}], '.format(annotation[element][0], annotation[element][1])
                
    return string_sorted_names
```

```python
%%time
%%capture info

tags_to_annotate = ['core', 'epi_nor', 'epi_mix', 'nor_mix', 'only_epi', 'only_nor', 'only_mix']

for tag in tags_to_annotate:
    print('###### {}'.format(tag))

    features_file = '{}.csv'.format(tag)
    features = pandas.read_csv(features_file, sep='\t', index_col=0)

    storing_mz = []; storing_rt = []; storing_annotation = []

    for i in range(len(features)):
        mz = features.iloc[i,:]['m/z']
        rt = features.iloc[i,:]['RT']

        print('about to annotate m/z {}; rt {}'.format(mz, rt))
        annotation = annotation_finder(mz, rt)
        print('final annotation: {}'.format(annotation))
        print()

        storing_mz.append(mz); storing_rt.append(rt); storing_annotation.append(annotation)

    newdf = pandas.DataFrame({'m/z':storing_mz,'RT':storing_rt, 'Annotation [IS | m/z diff]':storing_annotation})
    newdf.sort_values(by=['RT'], inplace=True)
    newdf.to_csv('{}.annotated.csv'.format(tag), sep='\t')
```
