```python
import pandas, numpy 
```

```python
import matplotlib, matplotlib.pyplot
matplotlib.rcParams.update({'font.size':20, 
                            'font.family':'sans-serif', 
                            'xtick.labelsize':16, 
                            'ytick.labelsize':16, 
                            'figure.figsize':(16*(2/3), 9*(2/3)), 
                            'axes.labelsize':20
                           })
```

```python
input_file = '../002.transcriptomics/supplemetary_information_table_SIT2.DETs.xlsx'
df_up = pandas.read_excel(input_file, sheet_name='5 uM | 4 h | up', index_col=0)
df_up.shape
```

```python
df_down = pandas.read_excel(input_file, sheet_name='5 uM | 4 h | down', index_col=0)
df_down.shape
```

```python
expression = pandas.concat([df_up, df_down])
expression.shape
```

```python
gene_expression = expression[expression['Biotype'] == 'protein_coding']
print(gene_expression.shape)
gene_expression
```

```python
input_file = '230117_Arnar.lfq.proteins_From_Gudrun.xlsx'
df = pandas.read_excel(input_file)
df.head()
```

```python
print('starting from {} genes'.format(gene_expression.shape[0]))
mapped = 0
for element in gene_expression['Gene name']:
    rule1 = df['Description'].str.contains('GN={} '.format(element))
    rule2 = df['Description'].str.contains('Fragment')
    subset = df[(rule1 == True) & (rule2 == False)]
    a = subset.shape[0]
    if a != 0:
        print('\t gene {} maps to {} proteins'.format(element, a))
        print('\t\t Description: ', gene_expression[gene_expression['Gene name'] == element]['Description'].values[0])
        mapped = mapped + 1

        # retreive mRNA fold-change
        mRNAfc = gene_expression[gene_expression['Gene name'] == element]['log2FC'].values[0]
        print('\t\t mRNAfc {}'.format(mRNAfc))
        
        # retrieve protein fold-change
        labels01 = subset.columns.values[[8, 9]]
        labels02 = subset.columns.values[[10, 11]]
        labels03 = subset.columns.values[[12, 13]]
        
        labels04 = subset.columns.values[[20, 21]]
        labels05 = subset.columns.values[[22, 23]]
        labels06 = subset.columns.values[[24, 25]]        
        
        adr1 = numpy.log2(subset.loc[:, labels01].median(axis=1) + 1)
        adr2 = numpy.log2(subset.loc[:, labels02].median(axis=1) + 1)
        adr3 = numpy.log2(subset.loc[:, labels03].median(axis=1) + 1)
        
        ctl1 = numpy.log2(subset.loc[:, labels04].median(axis=1) + 1)
        ctl2 = numpy.log2(subset.loc[:, labels05].median(axis=1) + 1)
        ctl3 = numpy.log2(subset.loc[:, labels06].median(axis=1) + 1)
        
        ptfc = numpy.median([adr1, adr2, adr3]) - numpy.median([ctl1, ctl2, ctl3])
        #print()
        #print(adr1, adr2, adr3, ctl1, ctl2, ctl3)
        #print(numpy.median([adr1, adr2, adr3]))
        print('\t\t PT fc {}'.format(ptfc))

        ptthreshold = numpy.log2(115/100)

        the_color='black'
        if (mRNAfc >= 1) & (ptfc >= ptthreshold):
            the_color='tab:red'
        if (mRNAfc <= -1) & (ptfc <= -ptthreshold):
            the_color='tab:blue'

        if (mRNAfc >= 1) & (-ptthreshold <= ptfc <= ptthreshold):
            the_color='tab:green'
        if (mRNAfc <= -1) & (-ptthreshold <= ptfc <= ptthreshold):
            the_color='tab:orange'

        if (mRNAfc <= -1) & (ptfc >= ptthreshold):
            the_color='magenta'
        if (mRNAfc >= 1) & (ptfc <= -ptthreshold):
            the_color='c'
        
        matplotlib.pyplot.scatter(mRNAfc, ptfc, s=250, c=the_color, alpha=2/3, edgecolors='none')
        matplotlib.pyplot.text(mRNAfc+0.11, ptfc+0.11/5, element, fontsize=12, color=the_color)
    
print('mapped {} proteins'.format(mapped))

matplotlib.pyplot.xlabel('mRNA change [log2FC]')
matplotlib.pyplot.ylabel('Protein change [log2FC]')
matplotlib.pyplot.xlim([-5, 5])
matplotlib.pyplot.ylim([-1, 1])
matplotlib.pyplot.grid(ls=':', alpha=1/2)
matplotlib.pyplot.show()
```
